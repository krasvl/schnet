```python
import matplotlib.pyplot as plt

from torch.utils.data import  DataLoader

from gnn.src.data.dataset import InMemoQM9Dataset
```

```python
dataset = InMemoQM9Dataset('./records')
```

```python
loader = DataLoader(dataset)
size = [0]*40
for i, (z, r, t) in enumerate(loader):
    size[z.size(1)] += z.size(0)

plt.bar(range(40), size, align='center', alpha=0.5)
plt.xlabel('Размер структуры')
plt.ylabel('Число экземпляров')
```
